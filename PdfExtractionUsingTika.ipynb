```python
import tika
tika.initVM()
from tika import parser 
import logging
import os.path
```

```python
logging.basicConfig(level=os.environ.get("LOGLEVEL", "INFO"))
```

```python
base_path = os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath(''))))
print(base_path)
```

```python
parsed = parser.from_file(base_path + "cch_orientation_paper_-_final-public_version-converted.docx")
doc_content = parsed["content"].strip()
doc_content = os.linesep.join([s for s in doc_content.splitlines() if s])

print('PDF File Content: \n',doc_content)
print('\n Metadata: \n', parsed["metadata"])
```
